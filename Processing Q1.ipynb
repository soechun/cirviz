```python
result = []
for index, row in data.iterrows():
#     if index > 0:
#         break;
#     else:
    for author_info in row["authors"]:
#         print(author_info['name'])
        if(len(author_info['ids'])>1):
            print(author_info)
        if(len(author_info['ids'])>0):
            result.append((author_info['ids'][0], author_info['name'], int(row['year']),1))
```

```python
result[0]
```

```python
import numpy as np
```

```python
df = pd.DataFrame(np.array(result), columns = ['id','author','year','count'])
```

```python
df[['count']] = df[['count']].apply(pd.to_numeric)
```

```python
df1 = df.groupby(['id','author','year'], as_index=False)[["count"]].sum()
```
