```python
import pandas as pd
```

```python
data = pd.read_json('Mini-project-2-data.json',lines=True)
```

```python
data = data[data["venue"] == "ICSE"]
```

```python
by_year = data.drop(columns=['inCitations','outCitations'])
by_year = by_year[['year','title']].groupby('year',as_index=False)
```

```python
by_year.agg('count').reset_index().to_csv('trendline.csv',sep=',')
```
